```python
import dlt
from pyspark.sql.functions import col,expr

@dlt.view
def cdctable():
    return spark.readStream.format("delta").table("workspace.default.cdctable")

dlt.create_streaming_table("target_scd1")

dlt.apply_changes(target = "target_scd1",
                  source = "cdctable",
                  keys = ["userId"],
                  sequence_by = col("date_col"),
                  apply_as_deletes = expr("date_col IS NULL"),
                  except_column_list = ["date_col"],
                  stored_as_scd_type = 1)
```

```python
import dlt
from pyspark.sql.functions import col,expr

@dlt.view
def cdctable():
    return spark.readStream.format("delta").table("workspace.default.cdctable2")

dlt.create_streaming_table("target_scd2")

dlt.apply_changes(target = "target_scd2",
                  source = "cdctable",
                  keys = ["userId"],
                  sequence_by = col("sequence"),
                  apply_as_deletes = expr("sequence IS NULL"),
                  except_column_list = ["sequence"],
                  stored_as_scd_type = 2)
```
